```python
import random
import pandas as pd
```

```python
numbers = [i for i in range(100)]

data = pd.Series(random.choice(numbers) for _ in range(100))
data
```

```python
# Tìm số lớn nhất
# Tìm số bé nhất
# Số lượng số nhỏ hơn giá trị trung bình?
# Số lượng số lớn hơn giá trị trung bình?
# Số lần xuất hiện của các số trong Series?
# Tính số lượng số chẵn
# Tính số lượng số lẻ
# Tính số lượng số âm, số dương?
```

```python
_max = data.values[0]
_min = data.values[0]
_sum = 0
for i in data.values:
    _sum += i
    if i > _max:
        _max = i
    if i < _min:
        _min = i
                
print(f'Số lớn nhất là {_max}')
print(f'Số bé nhất là {_min}')
```

```python
average = _sum/(len(data.values))
print(average)
```

```python
# Tính số lượng số chẵn
print(f'Số lượng các số chẵn {len(data[(data%2==0)])}')
```

```python
# Tính số lượng số lẻ
print(f'Số lượng các số lẻ {len(data[(data%2!=0)])}')
```

```python
# Số lần xuất hiện của các số trong Series?
numbers = []
series = list(data.values)

for i in series:
    check = True
    for number in numbers:
        if i == number:
            check = False
            break
    
    if check:
        print (f'Số lần lặp lại của số {i} là {series.count(i)}')
        numbers.append(i)
```

```python
dicts = {i:0 for i in range(100)}

for i in data.values:
    if i in dicts.keys():
        dicts[i] +=1
        
result=sorted(dicts.items(), key = lambda x:x[1], reverse=True)
result = dict(result)
print(result)
```

```python
# Chia phổ dđiểm từ 0-10 thành 10 đoạn.
# Tính số lượng sinh viên có điểm từng môn trong các phổ trên
subdf = pd.DataFrame()
subjects = ['Toán', 'Lý', 'Hoá', 'Sinh', 'Văn', 'Sử', 'Địa', 'Tiếng Anh']
_range = [i for i in range(1,11,1)]
s = []
for i in _range:
    subdicts = {}
    subdicts["Range"] = f"{i-1}->{i}"
    for subject in subjects:
        subdicts[subject] = len(df[subject][df[subject] <= i])-len(df[subject][df[subject] <= i-1])
    s.append(subdicts)
subdf = pd.DataFrame(s)
subdf
```

```python
# Thêm cột điểm trung bình
import math
def _average(Toan,Ly,Hoa,Sinh,Van,Su,Dia,TiengAnh):
    return round((2*(Toan+Van+TiengAnh)+Ly+Hoa+Sinh+Su+Dia)/11,2)

subdicts = {}

for index in range(df['Mã học sinh'].count()):
    subdicts[_average(df.iloc[index][4],\
                      df.iloc[index][5],\
                      df.iloc[index][6],\
                      df.iloc[index][7],\
                      df.iloc[index][8],\
                      df.iloc[index][9],\
                      df.iloc[index][10],\
                      df.iloc[index][11])] = df.iloc[index][0]
    
df['Điểm trung bình'] = subdicts
df
```

```python
# Tính điểm các khối A, A1,B,C,D
df["Điểm khối A"] = df["Toán"]+df["Lý"]+df["Hoá"]
df["Điểm khối A1"] = df["Toán"]+df["Lý"]+df["Tiếng Anh"]
df["Điểm khối B"] = df["Toán"]+df["Hoá"]+df["Sinh"]
df["Điểm khối C"] = df["Văn"]+df["Sử"]+df["Địa"]
df["Điểm khối D"] = df["Toán"]+df["Văn"]+df["Tiếng Anh"]*2
df
```

```python
A = 23.5
A1 = 22
B = 21.5
C = 22.5
D = 27.7
df["Trúng tuyển khối A"] = df["Điểm khối A"]>=A
df["Trúng tuyển khối A1"] = df["Điểm khối A1"]>=A1
df["Trúng tuyển khối B"] = df["Điểm khối B"]>=B
df["Trúng tuyển khối C"] = df["Điểm khối C"]>=C
df["Trúng tuyển khối D"] = df["Điểm khối D"]>=D
df
```

```python
# So sánh điểm xét vào dđại học các khối
sdicts = {}
Khoi = ["A","A1","B","C","D"]
for i in Khoi:
    sdicts[i] = round(df[f'Điểm khối {i}'].mean(),2)
    
sdicts = sorted(sdicts.items(),key = lambda x:x[1], reverse=True)
sdicts
```

```python
# Thêm xếp loại học sinh
def Xeploai(Hocsinh):
    if Hocsinh[12] > 9:
        check = True
        for i in range(4,12,1):
            if Hocsinh[i]< 8:
                check = False
                break
        if check:
            return 'Xuất sắc'
        
    if Hocsinh[12] > 8:
        check = True
        for i in range(4,12,1):
            if Hocsinh[i]< 6.5:
                check = False
                break
        if check:
            return 'Giỏi'
        
    if Hocsinh[12] > 6.5:
        check = True
        for i in range(4,12,1):
            if Hocsinh[i]< 4.5:
                check = False
                break
        if check:
            return 'Khá'
        else:
            return 'Trung bình'
    else:
        return 'Trung bình'
    
_Xeploai = []
for index in range(df['Mã học sinh'].count()):
    _Xeploai.append(Xeploai(list(df.iloc[index].values)))

df["Xếp loại"] = pd.Series(_Xeploai)
df
```
