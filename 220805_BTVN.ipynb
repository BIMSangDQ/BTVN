```python
text = input("Nhập chuỗi:")

numbers = text.split()

sum = 0
for i in numbers:
    sum += int(i)
print(f"Tổng các chữ số đã nhập = {sum}")

arithmetic_Mean = sum/(len(numbers))
print(f"Trung bình cộng các chữ số đã nhập = {arithmetic_Mean}")

negative_number=0
positive_number=0

for i in range(len(numbers)):
    if int(numbers[i]) < 0:
        negative_number = int(numbers[i])
        break
        
for i in range(len(numbers)):
    if int(numbers[i]) < 0 and int(numbers[i]) > negative_number:
        negative_number = int(numbers[i])
print(f"số nguyên âm lớn nhất trong các chữ số đã nhập = {negative_number}")

for i in range(len(numbers)):
    if int(numbers[i]) > 0:
        positive_number = int(numbers[i])
        break
        
for i in range(len(numbers)):
    if int(numbers[i]) > 0 and int(numbers[i]) < positive_number:
        positive_number = int(numbers[i])
print(f"số nguyên dương bé nhất trong các chữ số đã nhập = {positive_number}")
```
