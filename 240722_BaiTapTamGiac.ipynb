```python
import math
import turtle

Xa = float(input('Nhập Xa= '))
Ya = float(input('Nhập Ya= '))
Xb = float(input('Nhập Xb= '))
Yb = float(input('Nhập Yb= '))
Xc = float(input('Nhập Xc= '))
Yc = float(input('Nhập Yc= '))


t=turtle.Turtle()
t.shape('turtle')

# Hàm di chuyển turtle không để lại đường đi
def Goto(x, y):
    t.penup()
    t.goto(x, y)
    t.pendown()

# Hàm vẽ vòng tròn tại vị trí tâm x,y và bán kính r
def Draw_Point(x,y,r,color):
    Goto(x+r,y)
    t.setheading(0)
    t.left(90)
    t.pencolor(color)
    t.fillcolor(color)
    t.begin_fill()
    t.circle(r)
    t.end_fill()

# Hàm vẽ line từ tọa độ 2 điểm
def Draw_Line2Point(X1,Y1,X2,Y2,color):
    Goto(X1,Y1)
    t.color(color)
    t.goto(X2,Y2)
    
#Hàm vẽ đường cao từ điểm thứ nhất đến điểm 2, 3
#Name: Tên điểm chân đường cao
def Draw_Height (x1,y1,Vertex1_Name,x2,y2,x3,y3,Name,color):
    #phương trình đường 23: dạng a1X+b1Y=c1
    a1= (y2-y3)
    b1= (x3-x2)
    c1= (y2-y3)*x2 + (x3-x2)*y2
    #Phương trình đường thẳng qua điểm 1 và vuông góc vs 23 dạng a2X+b2Y=c2
    a2 = b1
    b2 = -a1
    c2 = x1*b1-y1*a1
    
    #Giao điểm 2 đường
    Yh = (c1*a2-c2*a1)/(b1*a2-b2*a1)
    Xh = (c1-b1*Yh)/a1
    
    Goto(Xh,Yh)
    t.pencolor('black')
    t.write(Name,move=False, font=('monaco',20),align='right')
    Draw_Point(Xh,Yh,5,color)
    
    Draw_Line2Point(x1,y1,Xh,Yh,color)
    Draw_Line2Point(x2,y2,Xh,Yh,color)
    Draw_Line2Point(x3,y3,Xh,Yh,color)
    
    dH = math.sqrt((x1-Xh)**2 + (y1-Yh)**2)
    print(f'\nChân đường cao từ đỉnh {Vertex1_Name}: {Xh}; {Yh}')
    print(f'Chiều dài đường cao {Vertex1_Name}{Name}: {dH}')

#Hàm vẽ đường trung trực từ điểm thứ nhất đến điểm 2, 3
#Name: Tên điểm chân đường trung trực
def Draw_Midperpendicular(x1,y1,Vertex1_Name,x2,y2,x3,y3,Name,color):
    Xe = (x2+x3)/2
    Ye = (y2+y3)/2
    
    Goto(Xe,Ye)
    t.pencolor('black')
    t.write(Name,move=False, font=('monaco',20),align='left')
    Draw_Point(Xe,Ye,5,color)
    Draw_Line2Point(x1,y1,Xe,Ye,color)
    
    dE = math.sqrt((x1-Xe)**2 + (y1-Ye)**2)
    print(f'\nChân đường trung trực từ đỉnh {Vertex1_Name}: {Xe}; {Ye}')
    print(f'Chiều dài đường trung trực {Vertex1_Name}{Name}: {dE}')

t.hideturtle()
t.speed(10)

#Vẽ các điểm A,B,C
Draw_Point(Xa,Ya,5,'black')
t.write('A',move=False, font=('monaco',20),align='left')
Draw_Point(Xb,Yb,5,'black')
t.write('B',move=False, font=('monaco',20),align='left')
Draw_Point(Xc,Yc,5,'black')
t.write('C',move=False, font=('monaco',20),align='left')

#Vẽ các đường trung trực
Draw_Midperpendicular(Xa,Ya,'A',Xb,Yb,Xc,Yc,'D','green')
Draw_Midperpendicular(Xb,Yb,'B',Xa,Ya,Xc,Yc,'E','green')
Draw_Midperpendicular(Xc,Yc,'C',Xb,Yb,Xa,Ya,'F','green')

#Tọa độ trọng tâm ABC
XG =(Xa+Xb+Xc)/3
YG =(Ya+Yb+Yc)/3
print(f'\nTrọng tâm G: {XG}; {YG}')
Draw_Point(XG,YG,5,'black')
t.write('G',move=False, font=('monaco',20),align='left')

#Vẽ đường cao tam giác
Draw_Height(Xa,Ya,'A',Xb,Yb,Xc,Yc,'I','red')
Draw_Height(Xb,Yb,'B',Xa,Ya,Xc,Yc,'J','red')
Draw_Height(Xc,Yc,'C',Xb,Yb,Xa,Ya,'K','red')

#Tìm tọa độ trực tâm
    #phương trình đường cao đỉnh A: dạng a1X+b1Y=c1
a1 = (Xc-Xb)
b1 = -(Yb-Yc)
c1 = Xa*(Xc-Xb)-Ya*((Yb-Yc))
     #phương trình đường cao đỉnh B: dạng a2X+b2Y=c2
a2 = (Xc-Xa)
b2 = -(Ya-Yc)
c2 = Xb*(Xc-Xa)-Yb*((Ya-Yc))
    #Giao điểm 2 đường
Yh = (c1*a2-c2*a1)/(b1*a2-b2*a1)
Xh = (c1-b1*Yh)/a1
print(f'\nTrực tâm H: {Xh}; {Yh}')
Draw_Point(Xh,Yh,5,'black')
t.write('H',move=False, font=('monaco',20),align='right')

#Vẽ 3 cạnh tam giác
t.width(2)
t.pencolor('black')
Goto(Xa,Ya)
t.goto(Xb,Yb)
t.goto(Xc,Yc)
t.goto(Xa,Ya)
    
turtle.done()
```
