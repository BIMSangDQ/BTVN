```python
#add thêm điểm trung bình của mỗi học sinh
scores = [["Hoàng",4,5,6],["Tùng",6,3,8],["Hương",2,9,10]]

_average = lambda x,y,z: round((x+y+z)/3,2)

for student in scores:
    student.append(_average(student[1],student[2],student[3]))
    
print(scores)
print()

# Sắp xếp theo tên, điểm từng loại
sort_name = sorted(scores,key = lambda x: x[0])
print(sort_name)

sort_maths = sorted(scores,key = lambda x: x[1])
print(sort_maths)

sort_physical = sorted(scores,key = lambda x: x[2])
print(sort_physical)

sort_chemistry = sorted(scores,key = lambda x: x[3])
print(sort_chemistry)

sort_average = sorted(scores,key = lambda x: x[4])
print(sort_average)
print()

#Tìm học sinh có điểm trung bình cao nhất
max_average = 0
name = []
for student in scores:
    if student[4]> max_average:
        name.clear()
        name.append(student[0])
        max_average = student[4]
    elif student[4]== max_average:
        name.append(student[0])

print ("Các học sinh có điểm trung bình cao nhất:", name)
print()

#Tạo list mới có xếp loại
new_list = []
for student in scores:
    if student[4] > 8:
        new_list.append(student)
        new_list[len(new_list)-1].append("Giỏi")
    elif student[4] > 7:
        new_list.append(student)
        new_list[len(new_list)-1].append("Khá")
    elif student[4] > 5:
        new_list.append(student)
        new_list[len(new_list)-1].append("Trung bình")
    else:
        new_list.append(student)
        new_list[len(new_list)-1].append("Yếu")
        
print(new_list)
```

```python
matrix = ((3,  4,  5,  -9),
          (9,  0,  2,  10),
          (2,  5,  6,  2),
         (-3,  4,  2,  7))

_average = lambda a,b,c,d: (a+b+c+d)/4

#1. Tìm giá trị trung bình của mảng 2 chiều này
ar = []
for _tuple in matrix:
    ar.append(_average(_tuple[0],_tuple[1],_tuple[2],_tuple[3]))
    
print (f"Giá trị trung bình của mảng 2 chiều là: {_average(ar[0],ar[1],ar[2],ar[3])}")
print()

#2. Tìm giá trị lớn nhất của mảng nằm trên đường chéo chính
_max = matrix[0][0]
for i in range(1,len(matrix),1):
    if _max <  matrix[i][i]:
        _max = matrix[i][i]
print (f"Gía trị lớn nhất của đường chéo : {_max}")
print()

#3. Tìm giá trị lớn nhất các cột
for j in range(len(matrix)):
    _max1 = matrix[0][j]
    for i in range(len(matrix)):
        if _max1 <  matrix[i][j]:
            _max1 = matrix[i][j]
    print(f"Giá trị lớn nhất của cột {j+1} là: {_max1}")
```

```python
nhatky = ((6,10),
          (12,14),
          (13,18),
           (16,17))

def distinct_list (_list):
    """
    Hàm loại bỏ các phần tử trùng nhau trong list
    
    Parameter:
    _list: list truyền vào
    
    Return:
    - new_list : là list đã loại bỏ các phần tử trùng nhau
    """
    new_list = []
    for item in _list:
        TF = False
        for it in new_list:
            if item == it:
                TF = True
                break
            else:
                TF = False
        if TF == False:
            new_list.append(item)
    return new_list

ar = []
for ca in nhatky:
    for i in range(int(ca[0]*2),int((ca[1]+0.5)*2),1):
        ar.append(i)

_array = []
for i in ar:
    _array.append(i/2)
    
_array = sorted(_array)
_array = distinct_list(_array)
print(_array)

st = _array[0]
ed = 0
so_gio_lam_viec = 0
for i in range(1,len(_array)):
    if _array[i]-_array[i-1]>0.5:
        ed = _array[i-1]
        so_gio_lam_viec += ed-st
        st = _array[i]
    
    if i == len(_array)-1:
        ed = _array[i]
        so_gio_lam_viec += ed-st


print("số h làm việc:", so_gio_lam_viec)
```
