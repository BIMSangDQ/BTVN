```python
# Bài 1: Theo luật từ năm 2020 thuế suất thu nhập cá nhân được tính như sau:
#         Nếu phần thu nhập tính thuế/tháng đến 5 triệu thì thuế suất là 5%.
#         Trên 5 triệu đến 10 triệu thì thuế suất là 10%.
#         Trên 10 triệu đến 18 triệu là 15%.
#         Trên 18 triệu đến 32 triệu là 20%.
#         Trên 32 triệu đến 52 triệu là 25%.
#         Trên 52 triệu đến 80 triệu là 30%.
#         Trên 80 triệu là 35%. Tính:
# 2.	Thuế TNCN phải đóng khi biết tổng thu nhập và số người phụ thuộc.
# Biết rằng, mức giảm trừ gia cảnh được áp dụng cho bản thân người đóng thuế là 11 triệu/tháng.
# Mức áp dụng đối với người phụ thuộc là 4,4 triệu/tháng.
# Thu nhập tính thuế thu nhập cá nhân là thu nhập còn lại sau khi tổng thu nhập đã trừ đi các tổng giảm trừ gia cảnh.
tongThuNhap = int(input("Nhập tổng thu nhập (VNĐ): "))
soNguoiPhuThuoc = int(input("Số người phụ thuộc: "))

ThueSuat1 = 5/100
ThueSuat2 = 10/100
ThueSuat3 = 15/100
ThueSuat4 = 20/100
ThueSuat5 = 25/100
ThueSuat6 = 30/100
ThueSuat7 = 35/100

Muc1 = 5000000
Muc2 = 10000000
Muc3 = 18000000
Muc4 = 32000000
Muc5 = 52000000
Muc6 = 80000000

GiamTruGiaCanh = soNguoiPhuThuoc*4400000
MucDongThue = 11000000
ThueTNCN = 0

if tongThuNhap<0 or soNguoiPhuThuoc<0:
    print("Lỗi nhập liệu")
else:
    thuNhapDongThue = tongThuNhap - (MucDongThue+GiamTruGiaCanh)
    
    if thuNhapDongThue <=0:
        ThueTNCN = 0
    elif thuNhapDongThue < Muc1:
        ThueTNCN = thuNhapDongThue*ThueSuat1
    elif thuNhapDongThue < Muc2:    
        ThueTNCN = Muc1*ThueSuat1 +(thuNhapDongThue-Muc1)*ThueSuat2
    elif thuNhapDongThue < Muc3:    
        ThueTNCN = Muc1*ThueSuat1 + Muc2*ThueSuat2 +(thuNhapDongThue-Muc2)*ThueSuat3
    elif thuNhapDongThue < Muc4:    
        ThueTNCN = Muc1*ThueSuat1 + Muc2*ThueSuat2 + Muc3*ThueSuat3 +(thuNhapDongThue-Muc3)*ThueSuat4
    elif thuNhapDongThue < Muc5:    
        ThueTNCN = Muc1*ThueSuat1 + Muc2*ThueSuat2 + Muc3*ThueSuat3 + Muc4*ThueSuat4\
        +(thuNhapDongThue-Muc4)*ThueSuat5
    elif thuNhapDongThue < Muc6:    
        ThueTNCN = Muc1*ThueSuat1 + Muc2*ThueSuat2 + Muc3*ThueSuat3 + Muc4*ThueSuat4\
        + Muc5*ThueSuat5 + (thuNhapDongThue-Muc5)*ThueSuat6
    else:   
        ThueTNCN = Muc1*ThueSuat1 + Muc2*ThueSuat2 + Muc3*ThueSuat3 + Muc4*ThueSuat4\
        + Muc5*ThueSuat5 + Muc6*ThueSuat6 + (thuNhapDongThue-Muc6)*ThueSuat7
    
    print (f"Số tiền thuế TNCN phải đóng: {int(ThueTNCN)} VNĐ" if ThueTNCN>0 else "Không cần đóng thuế TNCN")
```

```python
import math
import turtle

Xp = float(input('Nhập Xp= '))
Yp = float(input('Nhập Yp= '))

OP = math.sqrt(Xp**2 + Yp**2)
textprint =""

#Tọa độ X <0 xét khoảng cách OP. OP>2 thì nằm ngoài
if Xp<=0:
    
    if (OP > 200):
        textprint = ("Điểm P nằm ngoài vùng D")
    elif (OP == 200):
        textprint =("Điểm P nằm trên đường bao của vùng D")
    else:
        textprint =("Điểm P nằm trong vùng D")

else:
    #Đưa về mp dương ktr 1 lần khỏi chia trường hợp.
    Ym = abs(Yp)
    
    #phương trình đường AB: A(0;200) B(600;0)
    a1= 200
    b1= 600
    c1= 600*200

    #phương trình đường OP': P'(Xp,Ym)
    a2= (-Ym)
    b2= (Xp)
    c2= 0

    #Giao điểm 2 đường là điểm H
    Yh = (c1*a2-c2*a1)/(b1*a2-b2*a1)
    Xh = (c1-b1*Yh)/a1
    
    OH = math.sqrt(Xh**2 + Yh**2)
    HP = math.sqrt((Xh-Xp)**2 + (Yh-Ym)**2)
    
    # H trùng với P
    if (HP == 0):
        textprint =("Điểm P nằm trên đường bao của vùng D")
    elif (round(OP,3) > round(OH,3)):
        textprint =("Điểm P nằm ngoài vùng D")
    else:
        textprint = ("Điểm P nằm trong vùng D")
    
    #print(round(OP,3),round(OH,3),round(HP,3))
    #Trả lại điểm H theo P
    Yh = -Yh if Yp<0 else Yh

t=turtle.Turtle()
t.shape('turtle')
t.speed(10)

# Hàm di chuyển turtle không để lại đường đi
def Goto(x, y):
    t.penup()
    t.goto(x, y)
    t.pendown()

# Hàm vẽ vòng tròn tại vị trí tâm x,y và bán kính r
def Draw_Point(x,y,r,color):
    Goto(x+r,y)
    t.setheading(0)
    t.left(90)
    t.fillcolor(color)
    t.begin_fill()
    t.circle(r)
    t.end_fill()
    
#Vẽ lại đề
Draw_Point(0,0,5,'black')
t.write("O",move=False, font=('monaco',20),align='left')

t.color('blue')
Goto(0, 200)
t.write(f"(0:200)",move=False, font=('monaco',10),align='left')
t.left (90)
t.circle(200,180)
t.write(f"(0:-200)",move=False, font=('monaco',10),align='left')
t.goto(600,0)
t.write(f"(600:0)",move=False, font=('monaco',10),align='left')
t.goto(0,200)


#Vẽ điểm P
Draw_Point(Xp,Yp,5,'black')
t.color('black')
t.write(f"P ({Xp}:{Yp})\n{textprint}",move=False, font=('monaco',15),align='left')
Goto

t.hideturtle()
turtle.done()
```
