```python
def distinct_list (_list):
    """
    Hàm loại bỏ các phần tử trùng nhau trong list
    
    Parameter:
    _list: list truyền vào
    
    Return:
    - new_list : là list đã loại bỏ các phần tử trùng nhau
    """
    new_list = []
    for item in _list:
        TF = False
        for it in new_list:
            if item == it:
                TF = True
                break
            else:
                TF = False
        if TF == False:
            new_list.append(item)
    
    return new_list

def union_list (list1, list2):
    """
    Hàm lấy list union của 2 list
    
    Parameter:
    - list1: list 1
    - list2: list 2
    Return:
    - new_list : union của 2 list
    """
    new_list = list1 + list2
    return distinct_list(new_list)

def intersec_list (list1, list2):
    """
    Hàm lấy list intersec của 2 list
    Parameter:
    - list1: list 1
    - list2: list 2
    Return:
    - new_list : là list gồm các phần tử giao nhau của 2 list
    """ 
    new_list=[]
    for item1 in list1:
        for item2 in list2:
            if item1 == item2:
                new_list.append(item1)
                break
    return distinct_list(new_list)

def except_list(list1,list2):
    """
    Hàm lấy các phần tử khác nhau của list1 so với list2
    
    Parameter:
    - list1: list 1
    - list2: list 2
    
    Return:
    - new_list : các phần tử của list 1 ko tồn tại trong list 2
    """
    new_list=[]
    for item1 in list1:
        TF = False
        for item2 in list2:
            if item1 == item2:
                TF = True
                break
        if TF == False:
            new_list.append(item1)
    return distinct_list(new_list)
```
