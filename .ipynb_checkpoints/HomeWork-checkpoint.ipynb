```python
def VeHinhBinhHanh(Xa,Ya,Xb,Yb,Xc,Yc):
    """Vẽ hình bình hành từ 3 điểm
    
        Parameters:
        Xa (int): Hoành độ điểm A
        Ya (int): Tung độ điểm A
        Xb (int): Hoành độ điểm B
        Yb (int): Tung độ điểm B
        Xc (int): Hoành độ điểm C
        Yc (int): Tung độ điểm C
        
        Returns:
        Hình vẽ hình bình hành
    """
    import turtle

    t=turtle.Turtle()
    t.shape('turtle')

    def go_to(x,y):
      t.penup()
      t.goto(x,y)
      t.pendown()

    def draw_Circle(r,rotate):
      t.setheading(0)
      t.left(rotate)
      t.circle(r)

    #Tính trung điểm 3 cạnh
    XmidAC = (Xa+Xc)/2
    YmidAC = (Ya+Yc)/2

    XmidAB = (Xa+Xb)/2
    YmidAB = (Ya+Yb)/2

    XmidBC = (Xc+Xb)/2
    YmidBC = (Yc+Yb)/2

    t.hideturtle()
    t.speed(10)

    go_to(Xa,Ya)
    t.goto(Xb,Yb)
    t.goto(Xc,Yc)
    t.goto(Xa,Ya)

    #Vẽ 3 điểm ABC
    go_to(Xa+2,Ya)
    t.fillcolor('black')
    t.begin_fill()
    draw_Circle(2,90)
    t.end_fill()

    go_to(Xb+2,Yb)
    t.fillcolor('black')
    t.begin_fill()
    draw_Circle(2,90)
    t.end_fill()

    go_to(Xc+2,Yc)
    t.fillcolor('black')
    t.begin_fill()
    draw_Circle(2,90)
    t.end_fill()

    #Lấy D đối xứng B qua AC
    Xd = XmidAC*2-Xb
    Yd = YmidAC*2-Yb

    t.pencolor('red')
    go_to(Xa,Ya)
    t.goto(Xd,Yd)
    t.goto(Xc,Yc)

    go_to(Xd+2,Yd)
    t.fillcolor('red')
    t.begin_fill()
    draw_Circle(2,90)
    t.end_fill()

    #Lấy E đối xứng C qua AB
    Xe = XmidAB*2-Xc
    Ye = YmidAB*2-Yc

    t.pencolor('green')
    go_to(Xa,Ya)
    t.goto(Xe,Ye)
    t.goto(Xb,Yb)

    go_to(Xe+2,Ye)
    t.fillcolor('green')
    t.begin_fill()
    draw_Circle(2,90)
    t.end_fill()

    #Lấy F đối xứng A qua BC
    Xf = XmidBC*2-Xa
    Yf = YmidBC*2-Ya

    t.pencolor('blue')
    go_to(Xb,Yb)
    t.goto(Xf,Yf)
    t.goto(Xc,Yc)

    go_to(Xf+2,Yf)
    t.fillcolor('blue')
    t.begin_fill()
    draw_Circle(2,90)
    t.end_fill()

    #end
    turtle.done()
```

```python
def KiemTraTuGiacLom(Xa,Ya,Xb,Yb,Xc,Yc,Xd,Yd):
    """Kiểm tra tứ giác nhập vào có là tứ giác lõm
    
        Parameters:
        Xa (int): Hoành độ điểm A
        Ya (int): Tung độ điểm A
        Xb (int): Hoành độ điểm B
        Yb (int): Tung độ điểm B
        Xc (int): Hoành độ điểm C
        Yc (int): Tung độ điểm C
        Xd (int): Hoành độ điểm D
        Yd (int): Tung độ điểm D
        
        Returns:
        bool: trả về kết quả true nếu tứ giác lõm và false nếu là tứ giác lồi
    """
    
    import math
    #phương trình đường AC: dạng a1X+b1Y=c2
    a1= (Ya-Yc)
    b1= (Xc-Xa)
    c1= (Ya-Yc)*Xa + (Xc-Xa)*Ya

    #phương trình đường BD: dạng a2X+b2Y=c2
    a2= (Yb-Yd)
    b2= (Xd-Xb)
    c2= (Yb-Yd)*Xb + (Xd-Xb)*Yb

    #Giao điểm 2 đường chéo là điểm H
    Yh = (c1*a2-c2*a1)/(b1*a2-b2*a1)
    Xh = (c1-b1*Yh)/a1

    #Tứ giác lõm khi giao điểm nằm ngoài 1 đoạn thẳng
    Lenght_AC = math.sqrt((Xa-Xc)**2 + (Ya-Yc)**2)
    Lenght_BD = math.sqrt((Xb-Xd)**2 + (Yb-Yd)**2)
    Lenght_AH = math.sqrt((Xa-Xh)**2 + (Ya-Yh)**2)
    Lenght_CH = math.sqrt((Xc-Xh)**2 + (Yc-Yh)**2)
    Lenght_BH = math.sqrt((Xb-Xh)**2 + (Yb-Yh)**2)
    Lenght_DH = math.sqrt((Xd-Xh)**2 + (Yd-Yh)**2)

    # Tứ giác lõm khi điểm H nằm ngoài tứ giác
    return (Lenght_BD < (Lenght_BH+Lenght_DH)) or (Lenght_AC < (Lenght_AH+Lenght_CH))
    
```

```python
def TinhDuongCaoHinhTamGiac(Xa,Ya,Xb,Yb,Xc,Yc):
    """Vẽ hình bình hành từ 3 điểm
    
        Parameters:
        Xa (int): Hoành độ điểm A
        Ya (int): Tung độ điểm A
        Xb (int): Hoành độ điểm B
        Yb (int): Tung độ điểm B
        Xc (int): Hoành độ điểm C
        Yc (int): Tung độ điểm C
        
        Returns:
        In ra chiều dài các đường cao, trung tuyến từ các đỉnh và hình vẽ
    """
    
    import math
    import turtle

    t=turtle.Turtle()
    t.shape('turtle')

    # Hàm di chuyển turtle không để lại đường đi
    def Goto(x, y):
        t.penup()
        t.goto(x, y)
        t.pendown()

    # Hàm vẽ vòng tròn tại vị trí tâm x,y và bán kính r
    def Draw_Point(x,y,r,color):
        Goto(x+r,y)
        t.setheading(0)
        t.left(90)
        t.pencolor(color)
        t.fillcolor(color)
        t.begin_fill()
        t.circle(r)
        t.end_fill()

    # Hàm vẽ line từ tọa độ 2 điểm
    def Draw_Line2Point(X1,Y1,X2,Y2,color):
        Goto(X1,Y1)
        t.color(color)
        t.goto(X2,Y2)

    #Hàm vẽ đường cao từ điểm thứ nhất đến điểm 2, 3
    #Name: Tên điểm chân đường cao
    def Draw_Height (x1,y1,Vertex1_Name,x2,y2,x3,y3,Name,color):
        #phương trình đường 23: dạng a1X+b1Y=c1
        a1= (y2-y3)
        b1= (x3-x2)
        c1= (y2-y3)*x2 + (x3-x2)*y2
        #Phương trình đường thẳng qua điểm 1 và vuông góc vs 23 dạng a2X+b2Y=c2
        a2 = b1
        b2 = -a1
        c2 = x1*b1-y1*a1

        #Giao điểm 2 đường
        Yh = (c1*a2-c2*a1)/(b1*a2-b2*a1)
        Xh = (c1-b1*Yh)/a1

        Goto(Xh,Yh)
        t.pencolor('black')
        t.write(Name,move=False, font=('monaco',20),align='right')
        Draw_Point(Xh,Yh,5,color)

        Draw_Line2Point(x1,y1,Xh,Yh,color)
        Draw_Line2Point(x2,y2,Xh,Yh,color)
        Draw_Line2Point(x3,y3,Xh,Yh,color)

        dH = math.sqrt((x1-Xh)**2 + (y1-Yh)**2)
        print(f'\nChân đường cao từ đỉnh {Vertex1_Name}: {Xh}; {Yh}')
        print(f'Chiều dài đường cao {Vertex1_Name}{Name}: {dH}')

    #Hàm vẽ đường trung trực từ điểm thứ nhất đến điểm 2, 3
    #Name: Tên điểm chân đường trung trực
    def Draw_Midperpendicular(x1,y1,Vertex1_Name,x2,y2,x3,y3,Name,color):
        Xe = (x2+x3)/2
        Ye = (y2+y3)/2

        Goto(Xe,Ye)
        t.pencolor('black')
        t.write(Name,move=False, font=('monaco',20),align='left')
        Draw_Point(Xe,Ye,5,color)
        Draw_Line2Point(x1,y1,Xe,Ye,color)

        dE = math.sqrt((x1-Xe)**2 + (y1-Ye)**2)
        print(f'\nChân đường trung trực từ đỉnh {Vertex1_Name}: {Xe}; {Ye}')
        print(f'Chiều dài đường trung trực {Vertex1_Name}{Name}: {dE}')

    t.hideturtle()
    t.speed(10)

    #Vẽ các điểm A,B,C
    Draw_Point(Xa,Ya,5,'black')
    t.write('A',move=False, font=('monaco',20),align='left')
    Draw_Point(Xb,Yb,5,'black')
    t.write('B',move=False, font=('monaco',20),align='left')
    Draw_Point(Xc,Yc,5,'black')
    t.write('C',move=False, font=('monaco',20),align='left')

    #Vẽ các đường trung trực
    Draw_Midperpendicular(Xa,Ya,'A',Xb,Yb,Xc,Yc,'D','green')
    Draw_Midperpendicular(Xb,Yb,'B',Xa,Ya,Xc,Yc,'E','green')
    Draw_Midperpendicular(Xc,Yc,'C',Xb,Yb,Xa,Ya,'F','green')

    #Tọa độ trọng tâm ABC
    XG =(Xa+Xb+Xc)/3
    YG =(Ya+Yb+Yc)/3
    print(f'\nTrọng tâm G: {XG}; {YG}')
    Draw_Point(XG,YG,5,'black')
    t.write('G',move=False, font=('monaco',20),align='left')

    #Vẽ đường cao tam giác
    Draw_Height(Xa,Ya,'A',Xb,Yb,Xc,Yc,'I','red')
    Draw_Height(Xb,Yb,'B',Xa,Ya,Xc,Yc,'J','red')
    Draw_Height(Xc,Yc,'C',Xb,Yb,Xa,Ya,'K','red')

    #Tìm tọa độ trực tâm
        #phương trình đường cao đỉnh A: dạng a1X+b1Y=c1
    a1 = (Xc-Xb)
    b1 = -(Yb-Yc)
    c1 = Xa*(Xc-Xb)-Ya*((Yb-Yc))
         #phương trình đường cao đỉnh B: dạng a2X+b2Y=c2
    a2 = (Xc-Xa)
    b2 = -(Ya-Yc)
    c2 = Xb*(Xc-Xa)-Yb*((Ya-Yc))
        #Giao điểm 2 đường
    Yh = (c1*a2-c2*a1)/(b1*a2-b2*a1)
    Xh = (c1-b1*Yh)/a1
    print(f'\nTrực tâm H: {Xh}; {Yh}')
    Draw_Point(Xh,Yh,5,'black')
    t.write('H',move=False, font=('monaco',20),align='right')

    #Vẽ 3 cạnh tam giác
    t.width(2)
    t.pencolor('black')
    Goto(Xa,Ya)
    t.goto(Xb,Yb)
    t.goto(Xc,Yc)
    t.goto(Xa,Ya)

    turtle.done()
```

```python
def TinhTongCacSoChanHoacLe(chars,_isEvenNumber):
    """Tính tổng các số chẵn hoặc lẻ
    
        Parameters:
        chars (str): Chuỗi nhập vào
        _isEvenNumber(bool): nhập True nếu tính tổng các số chẵn,
                                  False nếu tính tổng các số lẻ
        
        Returns:
        int: kết quả.
    """
    textInput = input("Nhập chuỗi bất kì:")
    if _isEvenNumber == True:
        sum_EvenNumber = 2*chars.count('2') + 4*chars.count('4') + 6*chars.count('6')\
        + 8*textInput.count('8')
        return sum_EvenNumber
    else:
        sum_OddNumber = 1*chars.count('1') + 3*chars.count('3') + 5*chars.count('5')\
        + 7*textInput.count('7') + 9*textInput.count('9')
        return sum_OddNumber
```

```python
def TinhTienDien(number):
    """Tính tiền điện phải trả hàng tháng
    
        Parameters:
        number (str): Số điện sử dụng
        
        Returns:
        int: Số tiền điện theo số sử dụng đã nhập.
    """
    Fee = 0

    UnitPrice_1 = 1678 
    UnitPrice_2 = 1734 
    UnitPrice_3 = 2014 
    UnitPrice_4 = 2536
    UnitPrice_5 = 2834
    UnitPrice_6 = 2927

    Range1 = 50
    Range2 = 100
    Range3 = 200
    Range4 = 300
    Range5 = 400


    if (number <0):
        Print("Lỗi nhập liệu!")
    else:
        if number <= Range1:
            Fee += number*UnitPrice_1
        elif number <= Range2:
            Fee += (Range1*UnitPrice_1 + (number - Range1)*UnitPrice_2)
        elif number <= Range3:
            Fee += (Range1*UnitPrice_1 + (Range2-Range1)*UnitPrice_2 + (number - Range2)*UnitPrice_3)
        elif number <= Range4:
            Fee += (Range1*UnitPrice_1 + (Range2-Range1)*UnitPrice_2 + (Range3-Range2)*UnitPrice_3 +\
                    (number - Range3)*UnitPrice_4)
        elif number <= Range5:
            Fee += (Range1*UnitPrice_1 + (Range2-Range1)*UnitPrice_2 + (Range3-Range2)*UnitPrice_3 +\
                    (Range4-Range3)*UnitPrice_4 + (number - Range4)*UnitPrice_5)
        else:
            Fee += (Range1*UnitPrice_1 + (Range2-Range1)*UnitPrice_2 + (Range3-Range2)*UnitPrice_3 +\
                    (Range4-Range3)*UnitPrice_4 + (Range5-Range4)*UnitPrice_5 + (number - Range5)*UnitPrice_6)
            
    return Fee
```

```python
def PhuongTrinhBac2(a,b,c):
    """Trả về chuỗi dạng đa thức bậc 2: ax^2 + bx + c
    
        Parameters:
        a (int): Tham số a
        b (int): Tham số b
        c (int): Tham số c
        
        Returns:
        str: đa thức bậc 2.
    """ 
    string_a = ""
    string_b = ""
    string_c = ""

    if a != 0:
        if a == 1:
            string_a = "x^2"
        elif a == -1:
            string_a = "-x^2"
        else:
            string_a = f"{a}x^2"

    if b != 0:
        if b == 1:
            string_b = " + x"
        elif b == -1:
            string_b = " - x"
        elif b > 0:
            string_b = f" + {b}x"
        else:
            string_b = f" - {abs(b)}x"

    if c != 0:
        if c>0:
            string_c = f" + {abs(c)}"
        else:
            string_c = f" - {abs(c)}"

    return f"{string_a}{string_b}{string_c}"
```

```python
def TinhThueTNCN(tongThuNhap,soNguoiPhuThuoc):
    """Tính thuế thu nhập cá nhân
    
        Parameters:
        tongThuNhap (int): Tổng thu nhập của người đó
        soNguoiPhuThuoc (int): Số người phụ thuộc của người đó
        
        Returns:
        int: Tiền Thuế thu nhập cá nhân
    """

    ThueSuat1 = 5/100
    ThueSuat2 = 10/100
    ThueSuat3 = 15/100
    ThueSuat4 = 20/100
    ThueSuat5 = 25/100
    ThueSuat6 = 30/100
    ThueSuat7 = 35/100

    Muc1 = 5000000
    Muc2 = 10000000
    Muc3 = 18000000
    Muc4 = 32000000
    Muc5 = 52000000
    Muc6 = 80000000

    GiamTruGiaCanh = soNguoiPhuThuoc*4400000
    MucDongThue = 11000000
    ThueTNCN = 0

    if tongThuNhap<0 or soNguoiPhuThuoc<0:
        print("Lỗi nhập liệu")
    else:
        thuNhapDongThue = tongThuNhap - (MucDongThue+GiamTruGiaCanh)

        if thuNhapDongThue <=0:
            ThueTNCN = 0
        elif thuNhapDongThue < Muc1:
            ThueTNCN = thuNhapDongThue*ThueSuat1
        elif thuNhapDongThue < Muc2:    
            ThueTNCN = Muc1*ThueSuat1 +(thuNhapDongThue-Muc1)*ThueSuat2
        elif thuNhapDongThue < Muc3:    
            ThueTNCN = Muc1*ThueSuat1 + Muc2*ThueSuat2 +(thuNhapDongThue-Muc2)*ThueSuat3
        elif thuNhapDongThue < Muc4:    
            ThueTNCN = Muc1*ThueSuat1 + Muc2*ThueSuat2 + Muc3*ThueSuat3 +(thuNhapDongThue-Muc3)*ThueSuat4
        elif thuNhapDongThue < Muc5:    
            ThueTNCN = Muc1*ThueSuat1 + Muc2*ThueSuat2 + Muc3*ThueSuat3 + Muc4*ThueSuat4\
            +(thuNhapDongThue-Muc4)*ThueSuat5
        elif thuNhapDongThue < Muc6:    
            ThueTNCN = Muc1*ThueSuat1 + Muc2*ThueSuat2 + Muc3*ThueSuat3 + Muc4*ThueSuat4\
            + Muc5*ThueSuat5 + (thuNhapDongThue-Muc5)*ThueSuat6
        else:   
            ThueTNCN = Muc1*ThueSuat1 + Muc2*ThueSuat2 + Muc3*ThueSuat3 + Muc4*ThueSuat4\
            + Muc5*ThueSuat5 + Muc6*ThueSuat6 + (thuNhapDongThue-Muc6)*ThueSuat7

        return ThueTNCN
```

```python
def NghiemGanDung(a,b, alpha):
    """Tìm nghiệm gần đúng của fx = e^x + sin(x)
        trong vùng [a,b] và sai số alpha
    
        Parameters:
        a (float): Cận dưới của vùng tìm nghiệm
        b (float): Cận dưới của vùng tìm nghiệm
        alpha (float): sai số cho phép
        
        Returns:
        float: nghiệm gần đúng của pt
    """
    import math

    def fx(x):
        return (math.e)**x + math.sin(x)

    delta = abs(a-b)
    x = (a+b)/2

    n=0

    while delta>alpha:
        x = (a+b)/2
        delta = abs(a-b)

        if fx(a)*fx(x) <0:
            a=a
            b=x
        elif fx(b)*fx(x) <0:
            a=x
            b=b
        n+=1
        
    return x
```

```python
def XuLyChuoiSo(chars,option = 0):
    """Xử lý chuỗi số đã nhập và trả về kết quả theo option
    
        Parameters:
        chars (str): Dãy số nhập vào
        option (int):
            Nhập 0 nếu tính tổng.
            Nhập 1 nếu tính trung bình cộng
            Nhập 2 nếu tìm số âm lớn nhất
            Nhập 3 nếu tìm số dương bé nhất
        
        Returns:
        int: kết quả trả theo option
    """
    
    numbers = chars.split()

    _sum = 0
    for i in numbers:
        _sum += int(i)

    return _sum if option == 0

    arithmetic_Mean = _sum/(len(numbers))
    return arithmetic_Mean if option == 1

    negative_number=0
    positive_number=0

    for i in range(len(numbers)):
        if int(numbers[i]) < 0:
            negative_number = int(numbers[i])
            break

    for i in range(len(numbers)):
        if int(numbers[i]) < 0 and int(numbers[i]) > negative_number:
            negative_number = int(numbers[i])
    return negative_number if option == 2

    for i in range(len(numbers)):
        if int(numbers[i]) > 0:
            positive_number = int(numbers[i])
            break

    for i in range(len(numbers)):
        if int(numbers[i]) > 0 and int(numbers[i]) < positive_number:
            positive_number = int(numbers[i])
    return positive_number if option == 3
```

```python
def VeBanCoTuong(_width):
    """Vẽ bàn cờ tướng với kích thước ô đưa vào
    
        Parameters:
        _width (int): Kích thước ô đơn vị
    """
    import turtle

    # Hàm di chuyển turtle không để lại đường đi
    def Goto(x, y):
        t.setheading(0)
        t.penup()
        t.goto(x, y)
        t.pendown()

    # align == 0: center
    # align == 1: left
    # align == 2: right   
    def Draw_position(x,y,_witdh,align):
        if (align == 0 or align ==1):
            Goto(x+_witdh/20, y+_witdh/20)
            t.left(90)
            t.forward(_witdh/10)
            t.backward(_witdh/10)
            t.right(90)
            t.forward(_witdh/10)

            Goto(x+_witdh/20, y-_witdh/20)
            t.right(90)
            t.forward(_witdh/10)
            t.backward(_witdh/10)
            t.left(90)
            t.forward(_witdh/10)

        if (align == 0 or align ==2):
            Goto(x-_witdh/20, y+_witdh/20)
            t.left(90)
            t.forward(_witdh/10)
            t.backward(_witdh/10)
            t.left(90)
            t.forward(_witdh/10)

            Goto(x-_witdh/20, y-_witdh/20)
            t.right(90)
            t.forward(_witdh/10)
            t.backward(_witdh/10)
            t.right(90)
            t.forward(_witdh/10)
            
    t=turtle.Turtle()
    t.speed(10)

    for i in range(-5,5,1):
        Goto(-4*_witdh, (i+0.5)*_witdh)
        t.forward(8*_witdh)

    for i in range(-4,5,1):
        Goto(i*_witdh, (-4.5)*_witdh)
        t.left(90)
        if i == -4 or i == 4:
            t.forward(9*_witdh)
            t.right(90)
        else:
            t.forward(4*_witdh)
            t.penup()
            t.forward(_witdh)
            t.pendown()
            t.forward(4*_witdh)
            t.right(90)

    Goto(-1*_witdh, (-4.5)*_witdh)
    t.goto(_witdh,(-2.5)*_witdh)
    t.backward(2*_witdh)
    t.goto(_witdh, (-4.5)*_witdh)

    Goto(-1*_witdh, (4.5)*_witdh)
    t.goto(_witdh,(2.5)*_witdh)
    t.backward(2*_witdh)
    t.goto(_witdh, (4.5)*_witdh)


    Draw_position(-3*_witdh,-2.5*_witdh,_witdh,0)
    Draw_position(3*_witdh,-2.5*_witdh,_witdh,0)
    Draw_position(-3*_witdh,2.5*_witdh,_witdh,0)
    Draw_position(3*_witdh,2.5*_witdh,_witdh,0)

    for i in range(-4,5,2):
        if (i ==-4):
            Draw_position(i*_witdh,-1.5*_witdh,_witdh,1)
            Draw_position(i*_witdh,1.5*_witdh,_witdh,1)
        elif (i == 4):
            Draw_position(i*_witdh,-1.5*_witdh,_witdh,2)
            Draw_position(i*_witdh,1.5*_witdh,_witdh,2)
        else:
            Draw_position(i*_witdh,-1.5*_witdh,_witdh,0)
            Draw_position(i*_witdh,1.5*_witdh,_witdh,0)

    t.hideturtle()
    turtle.done()
```

```python
def TinhTuanChanLe(month,year,option = 0):
    """Tính số tuần chẵn hoặc lẻ của tháng, năm nhập vào
    
        Parameters:
        month (int): tháng nhập vào
        year (int): năm nhập vào
        option (int): Nhập 0 nếu muốn đếm số tuần chẵn
                      Nhập 1 nếu muốn đếm số tuần lẻ
        
        Returns:
        int: Số tuần chẵn hoặc lẻ theo option
    """
    import datetime

    if month <=0 or year <=0 or month>12:
        print("Lỗi nhập dữ liệu")
    else:
        date1 = datetime.datetime(year, month, 1)
        if month <12:
            date2 = datetime.datetime(year, month+1, 1)
        else:
            date2 = datetime.datetime(year+1, 1, 1)

        _string = ""
        for dayOfMonth in range(1,(date2-date1).days+1,1):
            _string += datetime.datetime(year,month,dayOfMonth).strftime("%w")

        if (option ==0):
            #Tuần chẵn tính từ thứ 2 đến chủ nhật
            return _string.count("1234560")
        else:
            #Nếu tháng không bắt đầu là thứ 2 thì có tuần lẻ đầu tháng
            #Nếu tháng không kết thúc là chủ nhật thì có tuần lẻ cuối tháng
            Weeks = 0
            if _string[0] != '1':
                Weeks +=1
            if _string[-1] != '0':
                Weeks +=1
            return Weeks
```
