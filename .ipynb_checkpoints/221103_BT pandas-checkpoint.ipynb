# Bài 1

```python
import random
import pandas as pd
```

```python
numbers = [i for i in range(100)]

data = pd.Series(random.choice(numbers) for _ in range(100))
data
```

```python
# Tìm số lớn nhất
# Tìm số bé nhất
# Số lượng số nhỏ hơn giá trị trung bình?
# Số lượng số lớn hơn giá trị trung bình?
# Số lần xuất hiện của các số trong Series?
# Tính số lượng số chẵn
# Tính số lượng số lẻ
# Tính số lượng số âm, số dương?
```

```python
_max = data.values[0]
_min = data.values[0]
_sum = 0
for i in data.values:
    _sum += i
    if i > _max:
        _max = i
    if i < _min:
        _min = i
                
print(f'Số lớn nhất là {_max}')
print(f'Số bé nhất là {_min}')
```

```python
data.max()
```

```python
data.min()
```

```python
data.sum()
```

```python
average = _sum/(len(data.values))
print(average)
```

```python
# Giá trị trung bình
data.mean()
```

```python
print('Các số lớn hơn trung bình')
len(data[(data > average)])
```

```python
print('Các số nhỏ hơn trung bình')
len(data[(data < average)])
```

```python
# Tính số lượng số chẵn
print(f'Số lượng các số chẵn {len(data[(data%2==0)])}')
```

```python
# Tính số lượng số lẻ
print(f'Số lượng các số lẻ {len(data[(data%2!=0)])}')
```

```python
# Số lần xuất hiện của các số trong Series?
numbers = []
series = list(data.values)

for i in series:
    check = True
    for number in numbers:
        if i == number:
            check = False
            break
    
    if check:
        print (f'Số lần lặp lại của số {i} là {series.count(i)}')
        numbers.append(i)
```

```python
dicts = {i:0 for i in range(100)}

for i in data.values:
    if i in dicts.keys():
        dicts[i] +=1
        
result=sorted(dicts.items(), key = lambda x:x[1], reverse=True)
result = dict(result)
print(result)
```

```python
# Có bao nhiêu học sinh oở từng tỉnh thành tham gia?
dicts = {}

for a in df['Quê quán']:
    if a in dicts.keys():
        dicts[a] += 1
    else:
        dicts[a] = 1
        
print(dicts)
```

```python
# Học sinh lớn tuổi nhất
a = df[['Họ tên','Ngày sinh']]
a[a['Ngày sinh'] == a['Ngày sinh'].min()]
```

```python
# Học sinh lớn nhỏ nhất
a[a['Ngày sinh'] == a['Ngày sinh'].max()]
```
