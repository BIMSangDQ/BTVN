```python
import turtle

t=turtle.Turtle()
t.shape('turtle')

def go_to(x,y):
  t.penup()
  t.goto(x,y)
  t.pendown()
  
def draw_Circle(r,rotate):
  t.setheading(0)
  t.left(rotate)
  t.circle(r)

Xa = 100
Ya = 150

Xb = 0
Yb = 0

Xc = 70
Yc = -80

#Tính trung điểm 3 cạnh
XmidAC = (Xa+Xc)/2
YmidAC = (Ya+Yc)/2

XmidAB = (Xa+Xb)/2
YmidAB = (Ya+Yb)/2

XmidBC = (Xc+Xb)/2
YmidBC = (Yc+Yb)/2

t.hideturtle()
t.speed(10)

go_to(Xa,Ya)
t.goto(Xb,Yb)
t.goto(Xc,Yc)
t.goto(Xa,Ya)

#Vẽ 3 điểm ABC
go_to(Xa+2,Ya)
t.fillcolor('black')
t.begin_fill()
draw_Circle(2,90)
t.end_fill()

go_to(Xb+2,Yb)
t.fillcolor('black')
t.begin_fill()
draw_Circle(2,90)
t.end_fill()

go_to(Xc+2,Yc)
t.fillcolor('black')
t.begin_fill()
draw_Circle(2,90)
t.end_fill()

#Lấy D đối xứng B qua AC
Xd = XmidAC*2-Xb
Yd = YmidAC*2-Yb

t.pencolor('red')
go_to(Xa,Ya)
t.goto(Xd,Yd)
t.goto(Xc,Yc)

go_to(Xd+2,Yd)
t.fillcolor('red')
t.begin_fill()
draw_Circle(2,90)
t.end_fill()

#Lấy E đối xứng C qua AB
Xe = XmidAB*2-Xc
Ye = YmidAB*2-Yc

t.pencolor('green')
go_to(Xa,Ya)
t.goto(Xe,Ye)
t.goto(Xb,Yb)

go_to(Xe+2,Ye)
t.fillcolor('green')
t.begin_fill()
draw_Circle(2,90)
t.end_fill()

#Lấy F đối xứng A qua BC
Xf = XmidBC*2-Xa
Yf = YmidBC*2-Ya

t.pencolor('blue')
go_to(Xb,Yb)
t.goto(Xf,Yf)
t.goto(Xc,Yc)

go_to(Xf+2,Yf)
t.fillcolor('blue')
t.begin_fill()
draw_Circle(2,90)
t.end_fill()

#end
turtle.done()
```
