##### 2. Kiểm tra dữ liệu khuyết thiếu

```python
population.isna()
```

```python
population = population.dropna()
```

```python
sub_population = population.drop(columns='Country')
sub_population
```

```python
from scipy import stats

z = stats.zscore(sub_population)
z
```

```python
from sklearn.preprocessing import StandardScaler

s_scaler = StandardScaler()
df_s = s_scaler.fit_transform(z)

df_s = pd.DataFrame(df_s,columns=['Population(millions)','Births per 1000 population','Deaths per 1000 populationi',
'Infant mortality rate','Urban Population (%)','Density of population'])
df_s
```

```python
import seaborn as sns
sns.boxplot(x = df_s['Population(millions)'])
```
