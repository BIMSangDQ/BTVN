### 3.Tìm thông tin ngân hàng có số lần giao dịch gửi tiền nhiều nhất trong ngày, trong tháng

```python
df['DATE'] = df['DAY'].astype(str) + "-" + df['MONTH'].astype(str)+ "-" + df['YEAR'].astype(str)
```

#### 3.1 giao dịch gửi tiền nhiều nhất trong ngày

```python
subdf = pd.DataFrame(df.pivot_table(values='MONEY',\
                index = 'DATE',
                aggfunc=max))

list_ = []
for index in range(subdf['MONEY'].count()):
    bank = df[(df['DATE'].str.contains(subdf.iloc[index].name)) & (df['MONEY'] == subdf.iloc[index]['MONEY'])]['FROM_CODE']
    list_.append(bank.iloc[0])

    
subdf['FROM_CODE'] = list_
subdf
```

#### 3.2 giao dịch gửi tiền nhiều nhất trong tháng

```python
subdf_month = pd.DataFrame(df.pivot_table(values='MONEY',\
                index = ['MONTH','FROM_CODE'],
                aggfunc=sum))


list_ = []

months = [i for i in range(1,13,1)]
for month in months:
    _max = 0
    i = 0
    dict_Month = {}
    dict_bank = {}
    dict_money = {}
    for index in range(subdf_month['MONEY'].count()):
        if subdf_month.iloc[index].name[0] == month and subdf_month.iloc[index]['MONEY']> _max:
            _max = subdf_month.iloc[index]['MONEY']
            i = index
    dict_bank['MONTH'] = int(month)
    dict_bank['FROM_CODE'] = subdf_month.iloc[i].name[1]
    dict_bank['MONEY'] = _max
    list_.append(dict_bank)
    list_.append(dict_Month)
    list_.append(dict_money)

subdf_month = pd.DataFrame(list_)
subdf_month.groupby(['MONTH','FROM_CODE'])['MONEY'].max()
```

##### => Giao dịch gửi tiền nhiều nhất trong tháng thường là ở ngân hàng BIDV, và tháng 1 không có giao dịch gửi

### 5.Tìm thông tin cặp ngân hàng nào có số lần giao dịch qua lại nhiều nhất

```python
ListBankBank = []
_dict = {}
for index in range(df['FROM_CODE'].count()):
    if ((df.iloc[index]['FROM_CODE']+'-'+df.iloc[index]['TO_CODE']) in ListBankBank) or ((df.iloc[index]['TO_CODE']+'-'+df.iloc[index]['FROM_CODE']) in ListBankBank):
        continue
    else:
        ListBankBank.append(df.iloc[index]['FROM_CODE']+'-'+df.iloc[index]['TO_CODE'])

len(ListBankBank)




```

```python
_dict = {}
for BankBank in ListBankBank:
    Bank = BankBank.split('-')
    _sum = (df[((df['FROM_CODE'].str.contains(Bank[0])) & (df['TO_CODE'].str.contains(Bank[1]))) |\
            ((df['FROM_CODE'].str.contains(Bank[1])) & (df['TO_CODE'].str.contains(Bank[0])))]['TRANS'].sum())
    _dict[BankBank] = _sum
_dict

subdf_trans = pd.DataFrame([_dict])
subdf_trans = subdf_trans.transpose()
subdf_trans[subdf_trans[0] == subdf_trans[0].max()]
```

### 10.Tính tổng số giao dịch theo khung giờ của các ngân hàng

```python
def HOURS24(hours):
    buoi = hours.split(' ')[1]
    h = hours.split(':')[0]

    if buoi == "AM":
        return int(h)
    else:
        return int(h)+12
    
    
```

```python
list_ = []
for index in range(df['HOURS'].count()):
    list_.append(HOURS24(df.iloc[index]['HOURS']))
    
df['HOURS24'] = pd.Series(list_)
df['B'] =  df['HOURS'].str.split(" ").str[1]
```

```python
df.groupby('HOURS24')['TRANS'].sum()
```

```python
df.groupby('B')['TRANS'].sum()
```
