```python
import pandas as pd

births = pd.read_csv('Data/births.csv')
births
```

```python
births.info()
```

```python
births['decade'] = (births['year']//10)*10
births
```

```python
births.groupby(['decade','gender'])['births'].aggregate('sum').unstack()
```

```python
births.pivot_table('births',index='decade', columns = 'gender', aggfunc='sum')
```
