```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Từ 0 đến 10 chia 100 giá trị
x = np.linspace(0,10,100)
x
```

##### Biểu đồ đường

```python
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))
plt.show()
```

```python
plt.plot(x,np.sin(x),label = 'sin(x)')
plt.plot(x,np.cos(x),label = 'cos(x)')
plt.title('Hàm sin(x), cos(x)',fontdict={'fontsize':12,'color':'blue'})
plt.xlabel('x values')
plt.ylabel('sin(x), cos(x) values')
plt.legend()
plt.show()

# show kết thúc section, sau show() không ảnh hưởng nữa
```

```python
plt.plot(x,np.sin(x-0),color='blue') #Specify color by name
plt.plot(x,np.sin(x-1),color='g') #Short color code (rgbmyk)
plt.plot(x,np.sin(x-2),color='0.75') #GrayScale 0-1
plt.plot(x,np.sin(x-3),color='#FFDD44') #hex code 
plt.plot(x,np.sin(x-4),color=(1.0,0.2,0.3)) #RGB tuple, value 0 and 1
plt.plot(x,np.sin(x-5),color='chartreuse') #all HTML color names
```

```python
plt.plot(x,x+0,linestyle='solid')  #-
plt.plot(x,x+1,linestyle='dashed') #--
plt.plot(x,x+2,linestyle='dashdot') #-.
plt.plot(x,x+3,linestyle='dotted') #:
```

- Customize Scatter

```python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)

colors = rng.rand(100)
sizes = 1000*colors
# alpha độ trong suốt
# cmap : miền màu tương ứng
#marker: o x * s
plt.scatter(x,y,c=colors,marker='o',s=sizes,alpha=0.6,cmap='viridis')
plt.colorbar()
plt.show()
```

### 2.Mảng đồ thị

```python
students = pd.read_csv('Data/students.csv')
students.head()
```

##### Nhiều biểu đồ cùng loại trên cùng Axes

```python
plt.plot(students.index,students['Toán'],color = 'r',label = 'Toán')
plt.plot(students.index,students['Lý'],color = 'b',label = 'Lý')
plt.plot(students.index,students['Hoá'],color = 'g',label = 'Hoá')
fig = plt.gcf()
fig.set_size_inches(18.5,6.5)
plt.legend(fontsize = 14, loc = 'lower right')
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.show()
```

##### Nhiều biểu đồ cùng loại trên một Figure

```python
fig,ax = plt.subplots(2,1)
ax[0].plot(students.index,students['Toán'],color = 'r',label = 'Toán')
ax[0].plot(students.index,students['Lý'],color = 'b',label = 'Lý')
ax[0].plot(students.index,students['Hoá'],color = 'g',label = 'Hoá')
ax[0].set_ylabel('Khối A', fontsize = 14)
ax[0].legend(fontsize = 14, loc = 'lower right')

ax[1].plot(students.index,students['Văn'],color = 'r',label = 'Văn')
ax[1].plot(students.index,students['Sử'],color = 'b',label = 'Sử')
ax[1].plot(students.index,students['Địa'],color = 'g',label = 'Địa')
ax[1].set_ylabel('Khối C', fontsize = 14)
ax[1].legend(fontsize = 14, loc = 'lower right')

fig.set_size_inches(18.5,6.5)
plt.show()
```

##### Nhiều biểu đồ khác loại trên 1 Axes

```python
students['ĐTB'] = (students['Toán']+students['Lý']+students['Hoá'])/3
```

```python
plt.bar(students.index,students['ĐTB'],label = "ĐTB", width=0.5)
axes1 = plt.gca()

axes2 = axes1.twiny()
axes2.plot(students.index,students['Toán'],label = "Toán",color = 'r', linewidth=0.5, marker='o')
plt.legend(loc='upper right')
axes1.set_ylabel('Thang điểm 10', fontsize = 14)
plt.show()
```

##### Nhiều biểu đồ khác loại trên một Figure

```python
fig,ax = plt.subplots(2,1)
ax[0].plot(students.index,students['Toán'],color = 'r',label = 'Toán')
ax[0].plot(students.index,students['Lý'],color = 'b',label = 'Lý')
ax[0].plot(students.index,students['Hoá'],color = 'g',label = 'Hoá')
ax[0].set_ylabel('Khối A', fontsize = 14)
ax[0].legend(fontsize = 14, loc = 'lower right')

ax[1].bar(students.index,students['ĐTB'],label = "ĐTB", width=0.5)

fig.set_size_inches(18.5,6.5)
plt.show()
```

### 3.Trực quan hóa dữ liệu với seaborn

##### Giới thiệu

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
# Load bộ dữ liệu có sẵn trong seaborn

tips = sns.load_dataset("tips")
tips.head()
```

```python
# styling and themes in Seaborn

plt.scatter(tips.total_bill,tips.tip)
sns.set_style("whitegrid")
plt.show()
```

##### Biểu đồ xu hướng

```python
sns.lmplot(x='total_bill',y='tip',data=tips)
sns.set_style("whitegrid")
plt.show()
```

```python
sns.lmplot(x='total_bill',y='tip',hue="smoker",data=tips)
```

```python
# palette Bảng màu
sns.lmplot(x='total_bill',y='tip',hue="smoker",data=tips,markers=["o","x"],palette="Set1",legend=True)
```

```python
sns.lmplot(x='total_bill',y='tip',hue="smoker",data=tips,markers=["o","x"],palette="Set1",legend=True,col='smoker')
```

```python
sns.lmplot(x='total_bill',y='tip',hue="smoker",data=tips,markers=["o","x"],palette="Set1",legend=True,row='smoker')
```
