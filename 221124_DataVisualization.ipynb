## Data Visualization

- Phần không gian đặt biểu đồ: Figure

### 1.Các dạng biểu đồ

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Từ 0 đến 10 chia 100 giá trị
x = np.linspace(0,10,100)
x
```

##### Biểu đồ đường

```python
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))
plt.show()
```

```python
plt.plot(x,np.sin(x),label = 'sin(x)')
plt.plot(x,np.cos(x),label = 'cos(x)')
plt.title('Hàm sin(x), cos(x)',fontdict={'fontsize':12,'color':'blue'})
plt.xlabel('x values')
plt.ylabel('sin(x), cos(x) values')
plt.legend()
plt.show()

# show kết thúc section, sau show() không ảnh hưởng nữa
```

- Thay đổi cài đặt Axes, Axis

```python
# Get current axes
axes = plt.gca()
axes.set_xlabel('Giá trị đầu vào',color='red',fontsize = 14)
axes.set_ylabel('Giá trị đầu ra',color='green',fontsize = 14)
axes.tick_params(axis = 'x',color = 'blue', size = 12)
axes.tick_params(axis = 'y',color = 'black', size = 12)


plt.show()
```

```python
axes = plt.gca()
axes.set_xlabel('Giá trị đầu vào',color='red',fontsize = 14)
axes.set_ylabel('Giá trị đầu ra',color='green',fontsize = 14)
axes.tick_params(axis = 'x',color = 'blue', size = 6)
axes.tick_params(axis = 'y',color = 'black', size = 6)

plt.plot(x,np.sin(x),label = 'sin(x)')
plt.plot(x,np.cos(x),label = 'cos(x)')
plt.title('Hàm sin(x), cos(x)',fontdict={'fontsize':12,'color':'blue'})
plt.xlabel('x values')
plt.ylabel('sin(x), cos(x) values')
plt.legend()
plt.show()
```

- Adjusting line colors and style

```python
plt.plot(x,np.sin(x-0),color='blue') #Specify color by name
plt.plot(x,np.sin(x-1),color='g') #Short color code (rgbmyk)
plt.plot(x,np.sin(x-2),color='0.75') #GrayScale 0-1
plt.plot(x,np.sin(x-3),color='#FFDD44') #hex code 
plt.plot(x,np.sin(x-4),color=(1.0,0.2,0.3)) #RGB tuple, value 0 and 1
plt.plot(x,np.sin(x-5),color='chartreuse') #all HTML color names
```

```python
plt.plot(x,x+0,linestyle='solid')  #-
plt.plot(x,x+1,linestyle='dashed') #--
plt.plot(x,x+2,linestyle='dashdot') #-.
plt.plot(x,x+3,linestyle='dotted') #:
```

```python
plt.plot(x,x+0,linestyle='-')  #-
plt.plot(x,x+1,linestyle='--') #--
plt.plot(x,x+2,linestyle='-.') #-.
plt.plot(x,x+3,linestyle=':') #:
```

- Axes Limits

```python
plt.plot(x,np.sin(x),color = 'r',linestyle= '--')
plt.xlim(-1,11)
plt.ylim(-1.5,1.5)
```

```python
# reverse order
plt.plot(x,np.sin(x),color='r',linestyle = '--')
plt.xlim(-1,11)
plt.ylim(1.5,-1.5)
```

##### Scatter Plot

```python
x = np.linspace(0,10,30)
y = np.sin(x)
plt.scatter(x,y,marker='o')
```

- Customize Scatter

```python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)

colors = rng.rand(100)
sizes = 1000*colors
# alpha độ trong suốt
# cmap : miền màu tương ứng
#marker: o x * s
plt.scatter(x,y,c=colors,marker='o',s=sizes,alpha=0.6,cmap='viridis')
plt.colorbar()
plt.show()
```

##### Histograms

```python
data = np.random.normal(0,1,1000)
plt.hist(data)
plt.show()
```

```python
# bins: số khoảng
plt.hist(data,bins=30, alpha=0.5,histtype='stepfilled',color='steelblue',edgecolor='none')
```

```python
x1 = np.random.normal(0,0.8,1000)
x2 = np.random.normal(-2,1,1000)
x3 = np.random.normal(3,2,1000)
#Đóng gói nhiều thuộc tính dạng dict
kwargs = dict(histtype='stepfilled',alpha=0.3,bins=40)
plt.hist(x1,**kwargs)
plt.hist(x2,**kwargs)
plt.hist(x3,**kwargs)
plt.show()
```

##### Barchart

```python
x = ['2017','2018','2019','2020','2021']
y=[234,389,333,402,451]
plt.bar(x,y)
plt.title('Sản lượng vải xuất khẩu',fontsize = 14)
plt.xlabel('Năm',fontsize = 12)
plt.ylabel('Sản lượng (tấn)',fontsize = 12)
plt.show()
```

```python
students = pd.read_csv("Data/students.csv")
students
```

```python
x = students.columns[5:]
y = students.iloc[0,5:]
plt.bar(x,y)
plt.title(f'Điểm thi tốt nghiệp của sinh viên {students.iloc[0][1]}')
plt.xlabel('Môn thi')
plt.ylabel('Điểm')
plt.show()
```

- Customize barchart

```python
# Thay đổi chiều rộng các cột, mặc định là 0.8
plt.bar(x,y,width = 0.5)
```

```python
plt.bar(x,y,bottom=0)
```

```python
# Cài đặt một màu hoặc một danh sách màu cho các cột
plt.bar(x,y,color = ['red','green','blue'])
```

```python
# Cài đặt nhãn cho các cột
plt.bar(x,y,tick_label = ['1','2','3','4','5','6','7','8'])
```

##### Pie chart

```python
labels = ['Gạo','Gas','Thịt','Trứng','Sữa']
sizes = [433,292,231,110,20]

plt.pie(sizes,labels=labels)
plt.title('Sản lượng xuất khẩu năm 2021',fontsize = 14)
plt.show()
```

```python
under_five = students['Toán'][students['Toán']<=5].count()
```

```python
five_to_eight= students['Toán'][(students['Toán']<=8) &(students['Toán']>5)].count()
```

```python
greater_eight= students['Toán'][students['Toán']>8].count()
```

```python
scores = [under_five,five_to_eight,greater_eight]
labels = ['Kém','Khá','Giỏi']

plt.pie(scores,labels=labels)
plt.title('Phân loại điểm toán')
plt.show()
```

- Customize pie chart

```python
# Cài đặt màu
plt.pie(scores,labels=labels,colors = ['r','g','b'])
plt.show()
```

```python
# Gán % cho các thành phần
plt.pie(scores,labels=labels,autopct='%1.2f%%')
plt.show()
```

```python
# Tỷ lệ giữa tâm của mỗi phần và nhãn autopct != None
plt.pie(scores,labels=labels,autopct='%1.2f%%',pctdistance=0.4)
plt.show()
```

```python
# Vẽ bóng mờ cho các thành phần
plt.pie(scores,labels=labels,shadow=True)
plt.show()
```

```python
# Bán kính hình tròn
plt.pie(scores,labels=labels,radius=1)
plt.show()
```
