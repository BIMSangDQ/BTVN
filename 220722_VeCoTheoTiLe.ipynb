```python
#Nhập chiều dài vẽ cờ theo tỉ lệ
import turtle
import math

Lenght = int(input("Nhập chiều dài lá cờ muốn vẽ:"))
R = 12
B = 72
H = 48
d = math.sqrt(B**2+H**2)
Scale = Lenght/B

cosA = B/d 
gA = math.degrees(math.acos(cosA))
#print(gA)

def Goto(x, y):
    t.penup()
    t.goto(x, y)
    t.pendown()

def Draw_Arc(r,rad):
    t.circle(r,rad)
    
def Draw_Circle(r,pencolor,fillcolor):
    t.pencolor(pencolor)
    t.fillcolor(fillcolor)
    t.begin_fill()
    t.circle(r)
    t.end_fill()
    
def Draw_Rec(B,H,pencolor,fillcolor):
    t.pencolor(pencolor)
    t.fillcolor(fillcolor)
    t.begin_fill()
    t.forward(B)
    t.left(90)
    t.forward(H)
    t.left(90)
    t.forward(B)
    t.left(90)
    t.forward(H)
    t.left(90)
    t.end_fill()

def Draw_RecWithSpace(B,H,Space,pencolor,fillcolor):
    t.pencolor(pencolor)
    t.fillcolor(fillcolor)
    t.begin_fill()
    t.forward(B)
    t.left(90)
    t.forward(H)
    t.left(90)
    t.forward(B)
    t.left(90)
    t.forward(H)
    t.left(90)
    t.end_fill()
    
    t.forward((B-Space)/2)
    
    t.pencolor('white')
    t.fillcolor('white')
    t.begin_fill()
    t.forward(Space)
    t.left(90)
    t.forward(H)
    t.left(90)
    t.forward(Space)
    t.left(90)
    t.forward(H)
    t.left(90)
    t.end_fill()
    
    t.pencolor(pencolor)
    t.backward((B-Space)/2)
    
t=turtle.Turtle()
t.shape('turtle')
t.hideturtle()
#boundary
Goto(-B*Scale/2,-H*Scale/2)
Draw_Rec(B*Scale,H*Scale,'black','white')

#Circle
t.penup()
t.goto(0,0)
t.setheading(0)
t.right(gA)
t.forward(R*Scale)
t.left(90)
t.pendown()

Draw_Circle(R*Scale,'black','blue')

t.fillcolor('red')
t.begin_fill()
t.setheading(0)
t.right(gA)
t.left(90)
Draw_Arc(R*Scale,180)
Draw_Arc((R/2)*Scale,180)
t.left(180)
Draw_Arc((R/2)*Scale,-180)
t.end_fill()

#1
t.penup()
t.goto(0,0)
t.setheading(0)
t.left(180-gA)
t.forward(R*Scale*1.5)
t.left(90)
t.backward(R*Scale/2)
t.right(90)
t.forward(2*Scale)
t.left(90)
t.pendown()

Draw_Rec(12*Scale,2*Scale,'black','black')

t.penup()
t.right(90)
t.forward(3*Scale)
t.left(90)
t.pendown()

Draw_Rec(12*Scale,2*Scale,'black','black')

t.penup()
t.right(90)
t.forward(3*Scale)
t.left(90)
t.pendown()

Draw_Rec(12*Scale,2*Scale,'black','black')

#2
Goto(0,0)
t.penup()
t.setheading(0)
t.left(gA)
t.forward(R*Scale*1.5+2*Scale)
t.left(90)
t.backward(R*Scale/2)
t.pendown()

Draw_RecWithSpace(12*Scale,2*Scale,Scale,'black','black')

t.penup()
t.right(90)
t.forward(3*Scale)
t.left(90)
t.pendown()

Draw_Rec(12*Scale,2*Scale,'black','black')

t.penup()
t.right(90)
t.forward(3*Scale)
t.left(90)
t.pendown()

Draw_RecWithSpace(12*Scale,2*Scale,Scale,'black','black')

#3
Goto(0,0)
t.penup()
t.setheading(0)
t.right(gA)
t.forward(R*Scale*1.5+2*Scale)
t.left(90)
t.backward(R*Scale/2)
t.pendown()

Draw_RecWithSpace(12*Scale,2*Scale,Scale,'black','black')

t.penup()
t.right(90)
t.forward(3*Scale)
t.left(90)
t.pendown()

Draw_RecWithSpace(12*Scale,2*Scale,Scale,'black','black')

t.penup()
t.right(90)
t.forward(3*Scale)
t.left(90)
t.pendown()

Draw_RecWithSpace(12*Scale,2*Scale,Scale,'black','black')
#4
Goto(0,0)
t.penup()
t.setheading(0)
t.right(180-gA)
t.forward(R*Scale*1.5+2*Scale)
t.left(90)
t.backward(R*Scale/2)
t.pendown()

Draw_Rec(12*Scale,2*Scale,'black','black')

t.penup()
t.right(90)
t.forward(3*Scale)
t.left(90)
t.pendown()

Draw_RecWithSpace(12*Scale,2*Scale,Scale,'black','black')

t.penup()
t.right(90)
t.forward(3*Scale)
t.left(90)
t.pendown()

Draw_Rec(12*Scale,2*Scale,'black','black')
#end
turtle.done()
```
