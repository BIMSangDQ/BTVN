### Sắp xếp dữ liệu theo họ học sinh

```python
def TiengVietKhongDau(Name):
    Name.strip()

    string1 = "áàảãạăắằẳẵặâấầẩẫậ"
    string2 = "ÁÀẢÃẠĂẮẰẲẴẶÂẤẦẨẪẬ"
    string3 = "éèẻẽẹêếềểễệ"
    string4 = "ÉÈẺẼẸÊẾỀỂỄỆ"
    string5 = "óòỏõọôốồổỗộơớờởỡợ"
    string6 = "ÓÒỎÕỌÔỐỒỔỖỘƠỚỜỞỠỢ"
    string7 = "íìỉĩị"
    string8 = "ÍÌỈĨỊ"
    string9 = "úùủũụưứừửữự"
    string10 = "ÚÙỦŨỤƯỨỪỬỮỰ"
    string11 = "ýỳỷỹỵ"
    string12 = "ÝỲỶỸỴ"
    string13 = "đ"
    string14 = "Đ"

    newName = ""
    for char in Name:
        if char in string1:
            newName += 'a'
        elif char in string2:
            newName += 'A'
        elif char in string3:
            newName += 'e'
        elif char in string4:
            newName += 'E'
        elif char in string5:
            newName += 'o'
        elif char in string6:
            newName += 'O'
        elif char in string7:
            newName += 'i'
        elif char in string8:
            newName += 'I'
        elif char in string9:
            newName += 'u'
        elif char in string10:
            newName += 'U'
        elif char in string11:
            newName += 'y'
        elif char in string12:
            newName += 'Y'
        elif char in string13:
            newName += 'd'
        elif char in string14:
            newName += 'D'
        else:
            newName += char
    return newName
```

```python
students['Họ'] = students['Họ tên'].str.split(' ').str[0]
students['Tên'] = students['Họ tên'].str.split(' ').str[-1]

_listHo = []
_listTen = []
for index in range(students['Họ'].count()):
    _listHo.append(TiengVietKhongDau(students['Họ'].iloc[index]))
    _listTen.append(TiengVietKhongDau(students['Tên'].iloc[index]))
students['Ho'] = pd.Series(_listHo)
students['Ten'] = pd.Series(_listTen)

df = students.sort_values(by = ['Ho','Ten'], ascending=True)
df.drop(columns=['Ho','Ten','Họ','Tên'],inplace = True)
students.drop(columns=['Ho','Ten'],inplace = True)
df
```
